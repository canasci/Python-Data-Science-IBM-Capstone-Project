# Question 1, Problem Statement: What is the Top Complaint Type?

First, we import the required libraries and import the CSV file

```python
import pandas as pd

df_data_1 = pd.read_csv("NYC311_Data.csv")
df_data_1.head()
```

Let's remove the unnecessary index column

```python
df_data_1.drop(labels=df_data_1.columns.values[0], axis=1, inplace=True)
df_data_1.head()
```

Change column names and create a new dataframe called df_311 which is a copy of df_data_1

```python
headers = ["Created Date", "Unique Key", "Complaint Type", "Incident Zip",
           "Incident Address", "Street Name", "Address Type", "City",
           "Resolution Description", "Borough", "Latitude", "Longitude",
           "Closed Date", "Location Type", "Status"]

df_data_1.columns = headers
df_311 = df_data_1.iloc[:, :]
df_311.head()
```

Then, import the remaining libraries for data analysis and visualization

```python
from datetime import datetime
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

mpl.style.use('ggplot') # for ggplot-like style
```

Since we want to find the answer to the question: "Which type of complaint should the Department of Housing Preservation and Development of New York City focus on first?", let's investigate the complaint types 

```python
complaint_types = df_311['Complaint Type'].value_counts()
complaint_types.sort_values(axis=0, ascending=False, inplace=True)
complaint_types
```

Let's unite the type of these complaints as "heat/hot water" since they have become the same type of complaint through the years(as it can be read in the dataframe).

```python
def uniform_values(condition, replace_values_with, pandas_series):
    return np.where(condition, replace_values_with, pandas_series)
    
condition1 = (df_311["Complaint Type"]=='PAINT - PLASTER')   
df_311["Complaint Type"] = uniform_values(condition1, 'PAINT/PLASTER', df_311['Complaint Type'])

condition2 = (df_311["Complaint Type"]=='HEATING')
df_311["Complaint Type"] = uniform_values(condition2, 'HEAT/HOT WATER', df_311['Complaint Type'])

condition3 = (df_311["Complaint Type"]=='Electric')
df_311["Complaint Type"] = uniform_values(condition3, 'ELECTRIC', df_311['Complaint Type'])

condition4 = (df_311["Complaint Type"]=='Unsanitary Condition')
df_311["Complaint Type"] = uniform_values(condition4, 'UNSANITARY CONDITION', df_311['Complaint Type'])

condition5 = (df_311["Complaint Type"]=='General')
df_311["Complaint Type"] = uniform_values(condition5, 'GENERAL', df_311['Complaint Type'])

condition6 = (df_311["Complaint Type"]=='Safety')
df_311["Complaint Type"] = uniform_values(condition6, 'SAFETY', df_311['Complaint Type'])

condition7 = (df_311["Complaint Type"]=='Appliance')
df_311["Complaint Type"] = uniform_values(condition7, 'APPLIANCE', df_311['Complaint Type'])

condition8 = (df_311["Complaint Type"]=='Plumbing')
df_311["Complaint Type"] = uniform_values(condition8, 'PLUMBING', df_311['Complaint Type'])

condition9 = (df_311["Complaint Type"]=='Outside Building')
df_311["Complaint Type"] = uniform_values(condition9, 'OUTSIDE BUILDING', df_311['Complaint Type'])

condition10 = (df_311["Complaint Type"]=='Mold')
df_311["Complaint Type"] = uniform_values(condition10, 'MOLD', df_311['Complaint Type'])
```

```python
complaint_types = df_311['Complaint Type'].value_counts()
complaint_types.sort_values(axis=0, ascending=False, inplace=True)
complaint_types[0:20]
```

```python
df_311['Created Date'] = pd.to_datetime(df_311['Created Date'], format='%Y-%m-%dT%H:%M:%S')
df_311.info()
print("\n")
df_311["Created Date"].dtype
```

```python
df_311['Created Year'] = df_311['Created Date'].map(lambda x: x.strftime('%Y')).astype(int)
df_311['Created Month'] = df_311['Created Date'].map(lambda x: x.strftime('%m')).astype(int)
```

```python
complaint_types[0:20].plot(kind='barh', color='steelblue', figsize=(16, 12), fontsize=12, grid=None)

plt.xlabel('Number of Complaints')
plt.ylabel('Type of Complaints')
plt.title('Department of Housing Preservation and Development of New York City Complaints Types')

# annotate value labels to each country
for index, value in enumerate(complaint_types): 
    label = format(int(value), ',') # format int with commas
    
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value + 100, index - 0.10), color='black')

plt.show()
```

We see that the top amount of complaint is of type "heat/hot water" by far.

Now, we know that the top amount of complaint is of type 'HEAT/HOT WATER' but before we say that it is the one the department needs to focus on first, we need to investigate the amounts of complaints as a timeseries since 'HEAT/HOT WATER' complaints may reduce in summer(by intuition) and become less significant than other types.

```python
# winter based group
group_winter = df_311[df_311['Created Month'].isin([11, 12, 1])].groupby(['Created Year', 'Complaint Type']).count()['Unique Key'].unstack()
group_winter
```

```python
# summer based group
group_summer = df_311[df_311['Created Month'].isin([6, 7, 8])].groupby(['Created Year', 'Complaint Type']).count()['Unique Key'].unstack()
group_summer
```

```python
group_winter.plot(kind='bar', sort_columns=True, legend=True, figsize=(30, 15), colormap='tab20')
plt.title("Winter Season Based Complaint Types Between {0}-{1}".format(group_winter.index.values[0], group_winter.index.values[-1]))
plt.legend(loc='best')
plt.ylabel("Frequency")
```

```python
group_summer.plot(kind='bar', sort_columns=True, legend=True, figsize=(30, 15), colormap='tab20')
plt.title("Summer Season Based Complaint Types Between {0}-{1}".format(group_summer.index.values[0], group_summer.index.values[-1]))
plt.legend(loc='best')
plt.ylabel("Frequency")
```

In conclusion, we can say that 'heat/hot water' complaints should be the main focus of the HPD in winter seasons and in general. Here is the graph illustrating the total amount of complaints on a yearly basis (2010-2019)

```python
timeseries_group = df_311.groupby(['Created Year', 'Complaint Type']).count()['Unique Key'].unstack().plot(figsize=(30, 15), colormap='tab20')
timeseries_group
```
