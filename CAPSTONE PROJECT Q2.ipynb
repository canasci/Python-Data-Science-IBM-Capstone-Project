Import NYC 311 HPD data

```python
import pandas as pd

df_data_1 = pd.read_csv("NYC311_Data.csv")
df_data_1.head()
```

Drop the column named 'Unnamed' which is the same as the index.

```python
df_data_1.drop(labels=df_data_1.columns.values[0], axis=1, inplace=True)
df_data_1.head()
```

Changed the name of the columns.

```python
headers = ["Created Date", "Unique Key", "Complaint Type", "Incident Zip",
           "Incident Address", "Street Name", "Address Type", "City",
           "Resolution Description", "Borough", "Latitude", "Longitude",
           "Closed Date", "Location Type", "Status"]

df_data_1.columns = headers
df_311 = df_data_1.iloc[:, :]
df_311.head()
```

import the required libraries

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

mpl.style.use('ggplot') # for ggplot-like style
```

# Data Wrangling

```python
df_311.dtypes
```

```python
complaint_types = df_311['Complaint Type'].value_counts()
complaint_types.sort_values(axis=0, ascending=False, inplace=True)
complaint_types
```

Deal with the duplicate complaint types

```python
def uniform_values(condition, replace_values_with, pandas_series):
    return np.where(condition, replace_values_with, pandas_series)
    
condition1 = (df_311["Complaint Type"]=='PAINT - PLASTER')   
df_311["Complaint Type"] = uniform_values(condition1, 'PAINT/PLASTER', df_311['Complaint Type'])

condition2 = (df_311["Complaint Type"]=='HEATING')
df_311["Complaint Type"] = uniform_values(condition2, 'HEAT/HOT WATER', df_311['Complaint Type'])

condition3 = (df_311["Complaint Type"]=='Electric')
df_311["Complaint Type"] = uniform_values(condition3, 'ELECTRIC', df_311['Complaint Type'])

condition4 = (df_311["Complaint Type"]=='Unsanitary Condition')
df_311["Complaint Type"] = uniform_values(condition4, 'UNSANITARY CONDITION', df_311['Complaint Type'])

condition5 = (df_311["Complaint Type"]=='General')
df_311["Complaint Type"] = uniform_values(condition5, 'GENERAL', df_311['Complaint Type'])

condition6 = (df_311["Complaint Type"]=='Safety')
df_311["Complaint Type"] = uniform_values(condition6, 'SAFETY', df_311['Complaint Type'])

condition7 = (df_311["Complaint Type"]=='Appliance')
df_311["Complaint Type"] = uniform_values(condition7, 'APPLIANCE', df_311['Complaint Type'])

condition8 = (df_311["Complaint Type"]=='Plumbing')
df_311["Complaint Type"] = uniform_values(condition8, 'PLUMBING', df_311['Complaint Type'])

condition9 = (df_311["Complaint Type"]=='Outside Building')
df_311["Complaint Type"] = uniform_values(condition9, 'OUTSIDE BUILDING', df_311['Complaint Type'])

condition10 = (df_311["Complaint Type"]=='Mold')
df_311["Complaint Type"] = uniform_values(condition10, 'MOLD', df_311['Complaint Type'])
```

```python
complaint_types = df_311['Complaint Type'].value_counts()
complaint_types.sort_values(axis=0, ascending=False, inplace=True)
complaint_types
```

We concluded in the previous question that the HPD should focus on heat/hot water complaints, thus we filter out our dataframe so it only includes the issues regarding complaint of type heat/hot water.

```python
df_heat = df_311[df_311['Complaint Type']=='HEAT/HOT WATER']
df_heat.head()
```

Drop the unavailable data(N/A)

```python
df_area = df_heat.loc[:, ['City', 'Borough', 'Incident Zip', 'Complaint Type']].dropna()
df_area.count()
```

```python
df_area.sort_values('Incident Zip' ,inplace=True)
df_area
```

```python
df_borough = df_area[['Borough', 'Complaint Type']].groupby(['Borough'])
df_borough.head()
```

```python
df_number_complaints = df_area[['Borough', 'Complaint Type']].groupby(['Borough']).count()
df_number_complaints.rename(columns={'Complaint Type': 'Number of Complaints'}, inplace=True)
df_number_complaints
```

```python
ax = df_number_complaints.plot(kind='barh', figsize=(18, 9), color='steelblue')
plt.ylabel('BOROUGH')
plt.xlabel('Number of Complaints')
plt.title('HEAT/HOT WATER Complaints based on NYC Boroughs')

for index, value in enumerate(df_number_complaints['Number of Complaints']): 
    label = format(int(value), ',') # format int with commas
    print(label)
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value , index ), color='black')

plt.show()
```

In the graph, we see that the borough with highest number of complaints of type 'heat/hot water' is 'BRONX', then 'BROOKLYN' being the closest to it in terms of amounts of complaints. The HPD should focus on the borough of BRONX since the boroughs labeled as 'Unspecified' are not included.

We can divide the complaints into groups of zip codes.

```python
df_number_complaints_zip = df_area[['Incident Zip', 'Complaint Type']].groupby(['Incident Zip']).count()
df_number_complaints_zip.rename(columns={'Complaint Type': 'Number of Complaints'}, inplace=True)
df_number_complaints_zip.sort_values('Number of Complaints', inplace=True, ascending=False)
df_number_complaints_zip
```

Let's visualize the top 15 amount of complaints(of type 'heat/hot water') for the corresponding zip codes

```python
df_number_complaints_zip.iloc[0:15, :]
```

```python
df_number_complaints_zip.iloc[0:15, :].plot(kind='barh', figsize=(16, 8), color='steelblue')
plt.ylabel('Zip')
plt.xlabel('Number of Complaints')
plt.title('HEAT/HOT WATER Complaints based on NYC Zip Codes')

for index, value in enumerate(df_number_complaints_zip['Number of Complaints'].values[:15]): 
    label = format(int(value), ',') # format int with commas
    print(label)
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value , index ), color='black')

plt.show()
```
